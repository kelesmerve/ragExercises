```python
from PIL import Image
import requests
from IPython.display import Markdown, display
from google.genai import types
from google.genai.types import GenerateContentConfig

system_instruction = """
  Sen Google Gemini BÃ¼yÃ¼k Dil Modeli konusunda uzmansÄ±n.
  Gemini SDK konusunda sorulacak sorulara aÃ§Ä±k, net, anlaÅŸÄ±lÄ±r ve kÄ±sa cevaplar ver.
  CevaplarÄ±n LLM kullanarak uygulama geliÅŸtirmeye yeni baÅŸlayan programcÄ±lara yol gÃ¶stersin.
  En son Gemini SDK versiyonu olan 2025 iÃ§in geÃ§erli olan cevaplarÄ± oluÅŸtur.
"""

chat_config = types.GenerateContentConfig(
    system_instruction=system_instruction,
    temperature=0.2,
    top_p=0.7,
    top_k=20,
)

```

```python
system_instruction = """
You are an expert software developer and a helpful coding assistant.
You are able to generate high-quality code in any programming language.
"""
chat_config = types.GenerateContentConfig(
    system_instruction=system_instruction,
)
chat = client.chats.create(
    model=MODEL_ID,
    config=chat_config,
)
```

```python
from pydantic import TypeAdapter

#sohbet gecmÄ±sÄ± content lÄ±stelerÄ±nden olusan bÄ±r lÄ±stedÄ±r
#typeAdapter bu pydantÄ±c turlerÄ± arasÄ±nda donusum yapabÄ±lÄ±r
history_adapter = TypeAdapter(list[types.Content])

#oncekÄ± bolumden sohbet nesnesÄ± kullanÄ±lÄ±r
chat_history = chat.get_history()


#json lÄ±stesÄ±ne donusturulur, json dizisi dÄ±ske veya verÄ±yÄ± sakladÄ±gÄ±n bÄ±r yere kaydedÄ±lebÄ±lÄ±r
json_history = history_adapter.dump_json(chat_history)
```

```python
# jsonu tekrar pydantÄ±c semaya donusturur
history = history_adapter.validate_json(json_history)

#json gecmÄ±sÄ±nÄ± kullanarak yenÄ± bÄ±r sohbet oturumu baslatÄ±r
new_chat = client.chats.create(
    model=MODEL_ID,
    config=chat_config,
    history=history,
)
response = new_chat.send_message("Fonksiyonun adÄ± neydi?")
Markdown(response.text)
```

```python
from pydantic import BaseModel
import json

class Recipe(BaseModel):
  recipe_name: str
  recipe_description: str
  recipe_ingredients: list[str]

response = client.models.generate_content(
    model=MODEL_ID,
    contents="PopÃ¼ler bir kurabiye tarifi ve malzemelerini ver.",
    config=types.GenerateContentConfig(
        response_mime_type="application/json",  #yanÄ±t MIME tÃ¼rÃ¼
        response_schema=Recipe,  # yanÄ±t ÅŸemasÄ±
    ),
)
print(json.dumps(json.loads(response.text), indent=4)) #yanÄ±tÄ± json formatÄ±nda yazdÄ±r
```

```python
import time
import os
from google import genai
from google.genai import types
from google.colab import userdata

# API AnahtarÄ±nÄ± al
try:
    API_KEY = userdata.get('gemini_api_key').strip()
except Exception:
    API_KEY = os.environ.get("GEMINI_API_KEY", "").strip()

client = genai.Client(api_key=API_KEY)

def generate_controlled(prompt, model="gemini-2.5-flash", min_interval=6):
    """
    API'den cevap alma sÃ¼resini Ã¶lÃ§er ve yazar.
    ArdÄ±ndan RPM limitini aÅŸmamak iÃ§in toplam sÃ¼reyi 6 saniyeye tamamlar.
    """
    start_time = time.time()  # â±ï¸ Kronometreyi baÅŸlat

    try:
        # 1. API'ye isteÄŸi gÃ¶nder
        response = client.models.generate_content(
            model=model,
            contents=prompt,
            config=types.GenerateContentConfig(
                temperature=0.7,
            )
        )

        # 2. Cevap geldiÄŸi an sÃ¼reyi Ã¶lÃ§
        generation_time = time.time() - start_time
        print(f"   âš¡ Cevap Ã¼retme sÃ¼resi: {generation_time:.2f} saniye")

        # 3. GÃ¼venlik beklemesi hesapla
        # EÄŸer cevap 6 saniyeden hÄ±zlÄ± geldiyse, aradaki fark kadar bekle
        if generation_time < min_interval:
            wait_time = min_interval - generation_time
            print(f"   â˜• GÃ¼venli limit iÃ§in {wait_time:.2f} saniye bekleniyor...")
            time.sleep(wait_time)
        else:
            print("   âœ… Ä°ÅŸlem 6 saniyeden uzun sÃ¼rdÃ¼ÄŸÃ¼ iÃ§in beklemeye gerek kalmadÄ±.")

        return response.text

    except Exception as e:
        print(f"âŒ Hata oluÅŸtu: {e}")
        time.sleep(10) # Hata durumunda 10sn soÄŸuma sÃ¼resi
        return None

# --- TEST KODU ---

sorular = ["Python nedir?", "Docker ne iÅŸe yarar?", "Linux neden gÃ¼venlidir?"]

print(f"ğŸš€ KontrollÃ¼ dÃ¶ngÃ¼ baÅŸlÄ±yor (Hedef: Her tur min. 6 sn)...\n")

for i, soru in enumerate(sorular, 1):
    print(f"SORU {i}: {soru}")
    cevap = generate_controlled(soru)

    if cevap:
        # CevabÄ±n sadece ilk cÃ¼mlesini gÃ¶sterelim ki ekran dolmasÄ±n
        ilk_cumle = cevap.split('.')[0]
        print(f"ğŸ“ Ã–ZET: {ilk_cumle}...\n")
        print("-" * 50)
```
